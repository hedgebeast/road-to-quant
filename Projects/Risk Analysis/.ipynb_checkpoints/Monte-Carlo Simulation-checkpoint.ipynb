```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import datetime as dt
import yfinance as yf
```

```python
num_of_sim = 100

def monte_carlo(S0, mu, sigma, N=252):
    
    
    result = []
    
    for _ in range(num_of_sim):
        
        prices = [S0]
        # daily step size
        for _ in range(N):

            stock_price = prices[-1] * np.exp((mu - 0.5*sigma**2) + sigma*np.random.normal()) 

            prices.append(stock_price)

        result.append(prices)
        
    simulation_df = pd.DataFrame(result)
    
    return simulation_df.T
```

```python
MC_paths = monte_carlo(50, 0.0002, 0.01, 1000)
# MC_paths['mean'] = MC_paths.mean(axis=1)
```

```python
MC_paths.plot(figsize=(20,8), lw=0.5, legend=False)
MC_paths.mean(axis=1).plot(lw=2, color="red", ls="--")
plt.grid(True, ls=":")
plt.xlim(0, 1000)
plt.show();
```

Stock prices follow a Geometric Process: $dS_T = \mu S_T dt + \sigma S_T dW$

$\implies S_T = S_0 e^{(\mu - \frac{1}{2}\sigma^2)T + \sigma \sqrt{T}Z}$ where Z~N(0,1)  

Hence, in Monte-Carlo simulation, we will generate a large amount of stock price estimates with this equation. Then, we will sort the stock prices in ascending order. 

If we need VaR with 99% confidence level, we will need the 1% lowest percentile $S_T$ in this series because there is a 1% chance that $S_0$ could fall to $S_T$ or less 

$VaR_{99\%} = S_0 - S_T$

```python
def download_stock(stock, start, end):
    data = {}
    ticker = yf.download(stock, start, end)
    data['Adj Close'] = ticker['Adj Close']
    
    return pd.DataFrame(data)
```

```python
class Var_MonteCarlo:
    
    def __init__(self, S, mu, sigma, c, T, _iter=100):
        
        self.S = S
        self.mu = mu
        self.sigma = sigma
        self.c = c
        self.T = T
        self.iterations = _iter
        
    def simulation(self): 
        
        rand = np.random.normal(0, 1, [1, self.iterations])
        
        #The random walk of our initial investment
        stock_price = self.S * np.exp((self.mu - 0.5*self.sigma**2)*self.T + self.sigma*np.sqrt(self.T)*rand)
        
        print(len(rand))
        
        #sort the stock prices to determine the percentile
        stock_price = np.sort(stock_price)     
        percentile = np.percentile(stock_price, (1-self.c)*100)
        
        return self.S - percentile #the loss at the specified confidence level
    
```

```python
S, c, T, _iter = 1e6, 0.95, 1, 100000 #_iter is the number of paths, T=1 (1-day VaR)

# Use historical data to approximate mean and std
start_date = dt.datetime(2020, 1, 1)
end_date = dt.datetime(2023, 11, 8)

tesla = download_stock('TSLA', start_date, end_date)
tesla['returns'] = tesla['Adj Close'].pct_change()

#We assume that daily returns to be normally distributed
mu = np.mean(tesla.returns)
sigma = np.std(tesla.returns)

model = Var_MonteCarlo(S, mu, sigma, c, T, _iter)
var = model.simulation()

print("VaR with Monte-Carlo Simulation: ${:.2f}".format(var))
```

# VaR Tools

# $\Delta VaR_i = \frac{\partial VaR_p}{\partial x_i} \hspace{0.3cm}$ where $x_i = w_i P_0$

$VaR_p = P_0|\mu_p T + \sigma_p \sqrt{T} Z| = -P_0(\mu_p T + \sigma_p \sqrt{T} Z) \hspace{0.3cm}$ where Z~N(0,1) 

### $\therefore \frac{\partial VaR_p}{\partial x_i} = \frac{\partial [-P_0(\mu_p T + \sigma_p \sqrt{T} Z)]}{\partial w_i P_0} = -\frac{\partial [\mu_p T + \sigma_p \sqrt{T} Z]}{\partial w_i}$

Note that $\mu_p = \sum_{j=1}^{n} w_j R_j = w_1 R_1 + ... + w_i R_i + ... + w_n R_n$ 

$\implies  -\frac{\partial [\mu_p T + \sigma_p \sqrt{T} Z]}{\partial w_i} =  -\frac{\partial [(\sum_{j=1}^{n} w_j R_j) T + \sigma_p \sqrt{T} Z]}{\partial w_i} = - R_i T - \frac{\partial \sigma_p}{\partial w_i} \sqrt{T} Z$

### The difficult part is deriving $\frac{\partial \sigma_p}{\partial w_i}$! 


In VaR Model Part 3, we established that $\sigma_{P}^2 = \sum_{i=1}^n \sum_{j=1}^n cov_{ij} w_i w_j = \sum_{j=1}^n w_j^2 \sigma_j^2 + 2 \sum_{k=1}^n \sum_{j \neq k}^n w_k w_j cov_{kj}$

## $\implies \frac{\partial \sigma_p^2}{\partial w_i} = \frac {\partial [\sum_{j=1}^n w_j^2 \sigma_j^2 + 2 \sum_{k=1}^n \sum_{j \neq k}^n w_k w_j cov_{kj}]}{\partial w_i}$ 

We can make use of this and make use of partial differentiation e.g. $\frac{d y^2}{dx} = 2y \cdot \frac{\partial y}{\partial x} \implies \frac{\partial \sigma_p^2}{\partial w_i} = 2 \sigma_p \frac{\partial \sigma_p}{\partial w_i}$ 

$\therefore 2 \sigma_p \frac{\partial \sigma_p}{\partial w_i} = \frac {\partial [\sum_{j=1}^n w_j^2 \sigma_j^2 + 2 \sum_{k=1}^n \sum_{j \neq k}^n w_k w_j cov_{kj}]}{\partial w_i} \hspace{0.3cm} \equiv \hspace{0.3cm} \frac{\partial \sigma_p}{\partial w_i} = \frac {\partial [\sum_{j=1}^n w_j^2 \sigma_j^2 + 2 \sum_{k=1}^n \sum_{j \neq k}^n w_k w_j cov_{kj}]}{2 \sigma_p \partial w_i}$ 

$\implies \frac{\partial \sigma_p}{\partial w_i} = \frac{1}{2 \sigma_p} \cdot 2w_i \sigma_i^2 + 2 \sum_{k \neq i} w_k cov_{ki} = \frac{1}{\sigma_p} \cdot \sum w_k cov_{ki}$ 


Recall that cov(x+y, z) = cov(x, z) + cov(y, z), we can apply it to $\sum w_k cov_{ki}$. 

$\implies \sum w_k cov_{ki} = cov(w_1 R_1, R_i) + ... + cov(w_n R_n, R_i) = cov(\sum w_k R_k, R_i) = cov(R_p, R_i)$

## HENCE, $\frac{\partial \sigma_p}{\partial w_i} = \frac{1}{\sigma_p} \cdot cov(R_i, R_p) \hspace{0.5cm}$ and thus,  $\hspace{0.5cm} \frac{\partial VaR_p}{\partial x_i} = - R_i T - \frac{1}{\sigma_p} \cdot cov(R_i, R_p) \sqrt{T} Z$

### $\frac{\partial VaR_p}{\partial x_i} = - R_i T - \frac{cov(R_i, R_p)}{\sigma_p} \cdot \sqrt{T} Z$ 

and we know that $\beta_i = \frac {cov_{ip}}{\sigma_p^2} \hspace{0.3cm} \equiv \hspace{0.3cm} \frac {cov_{ip}}{\sigma_p} = \beta_i \sigma_p$ 

$\therefore \Delta VaR_i = - R_i T - \beta_i \sigma_p \cdot \sqrt{T} Z$

### Incremental Va$R_i \approx (\Delta VaR_i) \times a$, where a is the extra position

Va$R_p = \sum$Component Va$R_i$
