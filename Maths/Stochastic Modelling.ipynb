### (1) Normal (Arithmetic) Binomial Tree

$S_T = S_0 + \sum_{i=1}^n \sigma \sqrt{\frac{T}{n}} X_i$

Central Limit Thereom states  

$\lim_{n-> \infty} \frac {1}{\sqrt{n}} \sum_{i=1}^n X_i$ ~ N(0,1)

Hence, 

$S_T = S_0 + \sigma \sqrt{T} Z$ where Z~N(0,1)

### (2) Lognormal (Geometric) Binomial Tree (model log-price process)

$E[ln(S_T)] = ln(S_0) + (r- \frac{\sigma^2}{2})T$, this includes the drift term of interest rate with some volatility

$V[ln(S_T)] = \sigma^2T $

...

Hence, $ln(S_T) = ln(S_0) + (r-\frac{1}{2}\sigma^2)T +\sigma \sqrt{T} Z$ where Z~N(0,1)

$\implies S_T = S_0exp[(r-\frac{1}{2}\sigma^2)T + \sigma \sqrt{T}Z]$

### Martingales (DRIFTLESS) - under the risk-neutral measure, all relative price processes should be a martigale

Conditional Expectation

$S_n = \frac{1}{1+r} E_n^*[S_{n+1}]$ $\implies \frac{S_n}{(1+r)^n} = E_n^*[\frac{S_{n+1}}{(1+r)^{n+1}}]$

=> $M_n = E_n^*[M_{n+1}]$

$V_n = E_n^*[\frac{V_N}{(1+r)N-n}]$

### Brownian Motion

$W_t$ ~ $\sqrt{t}N(0,1)$ $\implies W_t$ ~ N(0,t) 

```python
def simulate_BM(steps, T):
#     paths = 2**steps
    paths = 500
    deltaT = T/steps
    t = np.linspace(0, T, steps+1)

    X = np.hstack([np.zeros((paths, 1)),
          np.random.randn(paths, steps)])
        
    return t, np.cumsum(np.sqrt(deltaT) * X, axis=1)
```

```python
plt.figure(figsize=(15,8))
t, x = simulate_BM(10000, 100.0)
plt.plot(t, x.T, lw=0.5, ls='dotted')
# plt.hlines(y = 0, xmin = 0, xmax = 2, colors='r')
plt.show()
```
